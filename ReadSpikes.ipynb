```python
##TT1 is a noise spike chosen to illustrate
#use ttfile = clean.ntt to see a clean recordings
#noise.ntt

ttfile = './noisey.ntt'

ttfile = './clean.ntt'
import numpy as np
with open(ttfile, 'rb') as f:
    data = f.read()
f.close()
```

```python

#if you want to examine the header, run the following
print(data[:1488])
```

```python
spikedata = data[16384:]
nevents = len(spikedata)//304
print(nevents)

ts = np.ndarray((nevents,), '<Q', spikedata, 0, (304,))
waveforms = np.ndarray((nevents,128), '<h', 
                       spikedata, 48, (304,2)).reshape(nevents,32,4).transpose()
```

```python
import random 
print(np.shape(waveforms))
from matplotlib import pyplot as plt

for i in range(20):
   fig, ax = plt.subplots()
   spike_event = random.randint(1,nevents)
   for j in range(4):
      plt.plot(waveforms[j,:,spike_event])
      plt.title("{}".format(spike_event))

   plt.show()

   #fig.savefig("noshort{}.png".format(i))
   #fig.close()
    
```
